Filling missing values - Imputation

```python
df3 = pd.read_csv("AB_NYC_2019.csv")
```

```python
df3.shape
```

```python
df3.head()
```

```python
df3.isna().sum()
```

Numeric Data (Use Mean)

```python
mean_value = df3['reviews_per_month'].mean()
df3['reviews_per_month'].fillna(mean_value,inplace =True)
```

```python
df3.head()
```

```python
df3[df3["number_of_reviews"]==0].head(2)
```

```python
df3.isna().sum()
```

Categorical Data(Use Mode values)

```python
df3["last_review"].fillna(df3["last_review"].value_counts().index[0])
```

Categorical Data(Use new category)

```python
df3["last_review"].fillna("Not Reviewed",inplace = True)
```

```python
df3.head()
```

```python
df3.isna().sum()
```

```python
df3["host_name"].fillna("Unknown",inplace = True)
```

```python
df3.isna().sum()
```

```python
df3["name"].fillna(method = "pad",inplace = True)
```
